```python
#------------------------- Acc -------------------------------
barWidth = 0.5
fig = plt.subplots(figsize =(6, 4)) 

bing_saetze = [prozent_acc_bing] 
deepl_saetze = [prozent_acc_deepl] 
google_saetze = [prozent_acc_google] 
gpt_saetze = [prozent_acc_gpt] 

brs1 = np.arange(len(bing_saetze)) 
brs2= [x + barWidth for x in brs1] 
brs3 = [x + barWidth for x in brs2] 
brs4 = [x + barWidth for x in brs3] 

plt.bar(brs1, bing_saetze, color ='y', width = barWidth, 
        edgecolor ='grey', label ='Bing/Microsoft') 
plt.bar(brs2, deepl_saetze, color ='black', width = barWidth, 
        edgecolor ='grey', label ='Deepl') 
plt.bar(brs3, google_saetze, color ='yellow', width = barWidth, 
        edgecolor ='grey', label ='Google') 
plt.bar(brs4, gpt_saetze, color ='grey', width = barWidth, 
        edgecolor ='grey', label ='GPT')

#plt.text(bing_saetze, deepl_saetze, google_saetze, gpt_saetze, wo="center", va="top")
plt.xlabel('Acc (Berufsbezeichnungen)') 
plt.ylabel('') 
plt.xticks([r + barWidth for r in range(len(bing_saetze))], ['Acc'])  

plt.legend()
plt.show() 
```

```python
#S-Wert

#------------------------- S-Wert -------------------------------
barWidth = 0.5
fig = plt.subplots(figsize =(6, 4)) 

S_st_nst_bing = [S_st_nst_bing] 
S_st_nst_deepl = [S_st_nst_deepl] 
S_st_nst_google = [S_st_nst_google] 
S_st_nst_gpt = [S_st_nst_gpt] 

brs12 = np.arange(len(S_st_nst_bing)) 
brs22= [x + barWidth for x in brs12] 
brs32 = [x + barWidth for x in brs22] 
brs42 = [x + barWidth for x in brs32] 

plt.bar(brs12, S_st_nst_bing, color ='y', width = barWidth, 
        edgecolor ='grey', label ='Bing/Microsoft') 
plt.bar(brs22, S_st_nst_deepl, color ='black', width = barWidth, 
        edgecolor ='grey', label ='Deepl') 
plt.bar(brs32, S_st_nst_google, color ='yellow', width = barWidth, 
        edgecolor ='grey', label ='Google') 
plt.bar(brs42, S_st_nst_gpt, color ='grey', width = barWidth, 
        edgecolor ='grey', label ='GPT')

plt.xlabel('S-Wert (Berufsbezeichnungen)') 
plt.ylabel('') 
plt.xticks([r + barWidth for r in range(len(S_st_nst_bing))], ['S'])  

plt.legend()
plt.show() 
```
